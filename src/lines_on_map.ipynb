```python
countries = [ dict(
        type = 'scattergeo',
        locationmode = 'country names',
        lon = df_countries['LONG'],
        lat = df_countries['LAT'],
        hoverinfo = 'text',
        text = df_countries['SHORT_NAME'],
        mode = 'markers',
        marker = dict( 
            size=2, 
            color='rgb(255, 0, 0)',
            line = dict(
                width=3,
                color='rgba(68, 68, 68, 0)'
            )
        ))]
        
asylum_paths = []
for i in range( len(df_2016) ):
    asylum_paths.append(
         dict(
             type = 'scattergeo',
#             locationmode = 'USA-states',
             lon = [ df_2016['origin_lon'][i], df_2016['asylum_lon'][i] ],
             lat = [ df_2016['origin_lat'][i], df_2016['asylum_lat'][i] ],
             mode = 'lines',
             line = dict(
                 width = 2,
                 color = 'red',
             ),
             opacity = float(df_2016['Value'][i])/float(df_2016['Value'].max()),
         )
     )
    
layout = dict(
        title = '2016 Refugee Asylum paths for Syria, Afghanistan, Iraq, and Serbia <br>(Hover for Country names)',
        showlegend = False, 
        geo = dict(
#             scope='north america',
            showframe = False,
            showcoastlines = True,
            projection = dict(type = 'Mercator'),
            showland = True,
            landcolor = 'rgb(243, 243, 243)',
            countrycolor = 'rgb(204, 204, 204)',
        ),
    )
    
fig = dict( data=asylum_paths + countries, layout=layout )
py.iplot( fig )
```
