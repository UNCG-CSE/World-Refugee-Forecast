```python
df_countries = df2
df_countries.head()


# df_flight_paths = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2011_february_aa_flight_paths.csv')
# df_flight_paths.head()

airports = [ dict(
        type = 'scattergeo',
        locationmode = 'country names',
        lon = df_countries['LONG'],
        lat = df_countries['LAT'],
        hoverinfo = 'text',
        text = df_countries['SHORT_NAME'],
        mode = 'markers',
        marker = dict( 
            size=2, 
            color='rgb(255, 0, 0)',
            line = dict(
                width=3,
                color='rgba(68, 68, 68, 0)'
            )
        ))]
        
asylum_paths = []
for i in range( len( df ) ):
    asylum_paths.append(
         dict(
             type = 'scattergeo',
#             locationmode = 'USA-states',
             lon = [ df['origin_lon'][i], df['asylum_lon'][i] ],
             lat = [ df['origin_lat'][i], df['asylum_lat'][i] ],
             mode = 'lines',
             line = dict(
                 width = 1,
                 color = 'red',
             ),
             opacity = float(df['Value'][i])/float(df['Value'].max()),
         )
     )
    
layout = dict(
        title = 'Feb. 2011 American Airline flight paths<br>(Hover for airport names)',
        showlegend = False, 
        geo = dict(
#             scope='north america',
            showframe = False,
            showcoastlines = True,
            projection = dict(type = 'Mercator'),
            showland = True,
            landcolor = 'rgb(243, 243, 243)',
            countrycolor = 'rgb(204, 204, 204)',
        ),
    )
    
fig = dict( data=airports, layout=layout )
py.iplot( fig, filename='d3-flight-paths' )
```
