### ISO ALPHA-3 Country codes 
The data set contains some aggregated data that does not match a known country code, so we need to filter those aggregates out.   
Another dataframe is created containing the official country codes and merged with the origional refugee dataframe to filter out the 'Country Code' values that do not match a known code.  

```python
# read in the codes
iso_3 = pd.read_excel("../data/ref_by_c_orig/country_code_web.xls",  header=1)
# rename column for merging 
iso_3=iso_3.rename(columns = {'ISO ALPHA-3 code':'Country Code'})
# get rid of redundant data
iso_3=iso_3.drop('Country or area name', axis=1)
iso_3.head(20)
# merge to filter codes
refugees = pd.merge(iso_3, refugees)
# save the dataframe 
#refugees.to_pickle("../data/ref_by_c_orig/refugee_pickle_2")
refugees = pd.read_pickle("../data/ref_by_c_orig/refugee_pickle_2")
refugees.head(20)
```

### Isolate, clean and sort the data from 1990 
Starting with the earliest year for which we have data, 1990, we will now create an isolated sub-frame to be cleaned of 'NaN' values and sorted.  

```python
# create the sub-frame
temp = pd.DataFrame(columns=['Country Code','country','1990']) 
temp['Country Code'] = refugees['Country Code']
temp['country'] = refugees['country']
temp['1990'] = refugees['1990']
# remove NaN values
temp = temp.dropna().sort_values('1990',ascending=False)
# save the sub-frame 
#temp.to_pickle('../data/ref_by_c_orig/1990_pickle')
ref90 = pd.read_pickle('../data/ref_by_c_orig/1990_pickle')
ref90.head(20)
```

### Choropleths
We will now use plotly to visualize the refugee data by country of origin on a world map.   
A Plotly Choropleth 'locations' attribute can take in the ISO ALPHA-3 Country codes and map our refugee numbers to the appropriate Country on the map. 

```python
data = [ dict(
        type = 'choropleth',
    
    # can take country codes 
        locations = ref90['Country Code'],
        z = ref90['1990'],
        text = ref90['country'],
        colorscale = [[0,"rgb(0, 89, 179)"],[0.8, "rgb(26, 140, 255)"],[1,"rgb(255, 255, 255)"]],
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick = False,
            title = 'Refugee Population<br>by Country of Origin'),
      ) ]

```

```python
layout = dict(
    title = '1990 Refugee Population by Country of Origin<br>Source:\
            <a href="https://data.worldbank.org/indicator/SM.POP.REFG.OR">\
            The World Bank</a>',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
)
```

```python
fig = dict( data=data, layout=layout )
py.iplot( fig, validate=False, filename='d3-world-map_1990' )
```

### Melt year columns into a row 

```python
df = pd.melt(refugees, id_vars=["country", "Country Code"], 
                  var_name="Year", value_name="Refugees")
# df.to_pickle('../data/ref_by_c_orig/melt_pickle')
refugees=pd.read_pickle('../data/ref_by_c_orig/melt_pickle')
refugees.head(20)
```

```python
# refugees = refugees.dropna()
```

```python
data = [dict(type='choropleth',
            locations = refugees['Country Code'],
            z=refugees['Refugees'],
            text = refugees['country'],
            colorscale = [[0,"rgb(0, 89, 179)"],[0.8, "rgb(26, 140, 255)"],[1,"rgb(255, 255, 255)"]],
            autocolorscale = False,
            reversescale = True,
            marker = dict(
                 line = dict (
                    color = 'rgb(180,180,180)',
                    width = 0.5
                ) 
             ),
            colorbar = dict(
            autotick = False,
            title = 'Refugee Population<br>by Country of Origin')
            )]
```

```python
for i in range(26):
    data.append(data[0].copy())
    data[-1]['z'] = data[0]['z'] * np.random.rand(*data[0]['z'].shape)
```

```python
steps = []
for i in range(len(data)):
    step = dict(method='restyle',
                args=['visible', [False] * len(data)],
                label='{}'.format(i + 1990))
    step['args'][1][i] = True
    steps.append(step)
```

```python
sliders = [dict(active=0,
                pad={"t": 1},
                steps=steps)] 
```

```python
layout = dict(
    title = 'Refugee Population by Country of Origin 1990-2016<br>Source:\
            <a href="https://data.worldbank.org/indicator/SM.POP.REFG.OR">\
            The World Bank</a>',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(type = 'Mercator')
    ),
    sliders=sliders)
```

```python
fig = dict(data=data, 
           layout=layout)
```

```python
py.iplot(fig, validate=False, filename='d3-world-map_origin_sliders')
```
