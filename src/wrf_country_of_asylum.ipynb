### Data Cleaning: 

```python
df = pd.read_csv("../data/ref_by_c_asylum/ref_by_c_asylum.csv", header = 2)                 

# re-naming a column 
#df=df.rename(columns = {'Country Name':'country'})
df.set_index('Country Name')

# # deleting various columns 
del df["Indicator Name"]
del df["Indicator Code"]

# # dropping empty rows 
df=df.dropna(subset=['1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001','2002',
                  '2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016'], 
                   how='all')

# dropping empty columns 
df=df.dropna(axis=1, how='all')
df.head(20)
```

```python
# read in the codes
iso_3 = pd.read_excel("../data/ref_by_c_asylum/country_code_web.xls",  header=1)
# rename column for merging 
iso_3=iso_3.rename(columns = {'ISO ALPHA-3 code':'Country Code'})
# get rid of redundant data
iso_3=iso_3.drop('Country or area name', axis=1)
iso_3.head(20)
# merge to filter codes
df = pd.merge(iso_3, df)
# save the dataframe 
#df.to_pickle("../data/ref_by_c_asylum/refugee_asy_pickle")
refugees = pd.read_pickle("../data/ref_by_c_asylum/refugee_asy_pickle")
refugees.head(20)
```

### Slider Animations with Choropleths

```python
refugees=pd.read_pickle("../data/ref_by_c_asylum/refugee_asy_pickle")
refugees.head(20)
```

### Melt year columns into a row

```python
df = pd.melt(refugees, id_vars=["Country Name", "Country Code"], 
                  var_name="Year", value_name="Refugees")
# df.to_pickle('../data/ref_by_c_asylum/melt_pickle')
refugees=pd.read_pickle('../data/ref_by_c_asylum/melt_pickle')
refugees.head(20)
```

```python
data = [dict(type='choropleth',
            locations = refugees['Country Code'],
            z=refugees['Refugees'],
            text = refugees['Country Name'],
            colorscale = [[0,"rgb(0, 89, 179)"],[0.8, "rgb(26, 140, 255)"],[1,"rgb(255, 255, 255)"]],
            autocolorscale = False,
            reversescale = True,
            marker = dict(
                 line = dict (
                    color = 'rgb(180,180,180)',
                    width = 0.5
                ) 
             ),
            colorbar = dict(
            autotick = False,
            title = 'Refugee Population<br>by Country of Asylum')
            )]
```

```python
for i in range(26):
    data.append(data[0].copy())
    data[-1]['z'] = data[0]['z'] * np.random.rand(*data[0]['z'].shape)
```

```python
steps = []
for i in range(len(data)):
    step = dict(method='restyle',
                args=['visible', [False] * len(data)],
                label='{}'.format(i + 1990))
    step['args'][1][i] = True
    steps.append(step)
```

```python
sliders = [dict(active=0,
                pad={"t": 1},
                steps=steps)] 
```

```python
layout = dict(
    title = 'Refugee Population by Country of Asylum 1990-2016<br>Source:\
            <a href="https://data.worldbank.org/indicator/SM.POP.REFG">\
            The World Bank</a>',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(type = 'Mercator')
    ),
    sliders=sliders)
```

```python
fig = dict(data=data, 
           layout=layout)
```

```python
py.iplot(fig, validate=False, filename='d3-world-map_asylum_sliders')
```
